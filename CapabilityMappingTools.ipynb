```python
import os
```

```python
model_path = '/Users/ylou/Desktop/SpringBoot/CoreCapabilityCommon/src/main/java/com/wyze/core/capability/common/model'
```

```python
files = os.listdir(model_path)
```

```python
capability_names = [file.split(".")[0] for file in files]
```

```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/ylou/Desktop/CamValue2Capability.csv", index_col = 0)
```

```python
df 
```

```python
capability_dict = {capability_name:{} for capability_name in capability_names}
```

```python
capability_dict
```

```python
for capability_name in capability_names:
    with open(model_path + "/" + capability_name + ".java", "r") as f:
        variable_flag = False
        for x in f:
            if variable_flag:
                if "private" in x:
                    words = x.split()
                    class_type = words[1]
                    property_name = words[2][:-1]
                    capability_dict[capability_name][property_name] = class_type
            else:
                if "capabilityType" in x:
                    variable_flag = True
```

```python
capability_dict
```

```python
for ucap in df.Capability.unique():
    if ucap != "nan":
        print(capability_dict[ucap])
```

```python
sorted_df = df.sort_values(by=['Capability', "Index"])
```

```python
sorted_df.fillna("null", inplace=True)
```

```python
sorted_df
```

```python
last_cap = "AudioAlarmSetting"
for index, row in sorted_df.iterrows():
    curr_cap = row.Capability
    if curr_cap != last_cap:
        print("    return " + last_cap + ".class;")
        last_cap = curr_cap
    case_str = "case \"" + str(index) +"\":"
    print(case_str)
```

```python
capability_dict
```

```python
property_type_dict = {}
```

```python
for ucap in sorted_df.Capability.unique():
    if ucap != "null":
        for k,v in capability_dict[ucap].items():
            if k not in property_type_dict:
                property_type_dict[k] = v
            elif property_type_dict[k] != v:
                print(ucap, k)
```

```python
sorted_ptype_dict = {k: v for k, v in sorted(property_type_dict.items(), key=lambda item: item[1])}
```

```python
sorted_ptype_dict
```

```python
last_v = 'Boolean'
for k,v in sorted_ptype_dict.items():
    if v != last_v:
        print("    return " + last_v + ".class;")
        last_v = v
    print("case \"" + str(k) + "\":")
print("    return " + last_v + ".class;")
```
